# 레드와인의 화확적 성분과 등급을 기록한 데이터


-출처: https://archive.ics.uci.edu/dataset/186/wine+quality

```python
import pandas as pd

redwine = pd.read_csv("winequality-red.csv", sep=';')
redwine.head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

redwine_X = redwine.iloc[:, :-1].to_numpy()
redwine_y = pd.get_dummies(redwine.iloc[:, -1]).to_numpy()

# train-test 분리
train_X, test_X, train_y, test_y = train_test_split(redwine_X, redwine_y, test_size=0.3, random_state=1)
train_X.shape, train_y.shape, test_X.shape, test_y.shape
```

# 문제 정의: 와인의 품질 속성을 분류하기
# 모형 설계

1. Sequential( )
- 레이어1: 뉴런 50, 활성화함수 relu
- 레이어2: 뉴런 50, 활성화함수 relu
- 레이어3: 뉴런 30, 활성화함수 relu
- 레이어 4 - 출력 6, softmax

2. compile( )
- 최적화: adam
- 손실함수: 'categorical_crossentropy'
- 평가지표: 'accuracy'

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# 1. 모델 정의
model = Sequential()
model.add(Input(shape=(11,))) #입력층 11개 피처 지정
model.add(Dense(65, activation='relu')) # 은닉층 레이어1
model.add(Dense(65, activation='relu')) # 은닉층 레이어2
model.add(Dense(30, activation='relu')) # 은닉층 레이어3
model.add(Dense(6, activation='softmax')) # 출력층 레이어4
model
```

```python
# 모델 설정 - 컴파일
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model
```

# 훈련

```python
# 모델 훈련
from tensorflow.keras.callbacks import EarlyStopping
# Early Stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

model.fit(train_X, train_y, epochs=500, batch_size=50, callbacks=[early_stopping], verbose=1)
```

# 평가

```python
pred = np.argmax(model.predict(test_X), axis=1)
pred
```

```python
# 실제값과 비교

np.argmax(test_y, axis=1)
```

```python
pd.crosstab(np.argmax(test_y, axis=1), pred, margins=True)
```

```python
loss, accuracy = model.evaluate(test_X, test_y)
loss, accuracy
```
